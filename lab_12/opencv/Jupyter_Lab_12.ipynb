### Ejercicio 4. Guarde una nueva imagen que contenga solo los canales rojo y verde

```python
import cv2
```

```python
img_cv = cv2.imread('Imagenes/playa.jpg') # trabaja con canales BGR
img_rg = img_cv.copy()

img_rg[:,:,0] = 0  # aqui se anula el canal Blue
cv2.imwrite('Imagenes/img_rg.png',img_rg)

cv2.imshow('red y green', img_rg) # visualizacion del resultado
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Ejercicio 5. Guarde una nueva imagen que muestre la imagen en escala de grises

```python
img_gris = cv2.imread('Imagenes/playa.jpg',0) # el segundo par√°metro en cero anula los colores
cv2.imwrite('Imagenes/img_gris.png',img_gris)

cv2.imshow('grises', img_gris) # visualizaci√≥n del resultado
cv2.waitKey(0)
cv2.destroyAllWindows()
```
